## 4.2 Limpieza de carcteres no deseados üôÖ:

Realizamos una funci√≥n para normalizar texto en columnas tipo string

```python
def normalizar_texto(df):
    for col in df.select_dtypes(include="object").columns:
      df[col]= (
          df[col].astype(str).str.strip().str.replace(r"[\u200b\t\r\n]", "", regex=True)
          .str.replace(" +", " ", regex=True).str.title()
          )
    return df
```

Realizamos una funci√≥n para normalizar las fechas en columnas tipo datetime

```python
def normalizar_fecha(df):
  for col in df.columns:
      if "fecha" in col.lower():
          df[col] = pd.to_datetime(df[col], errors="coerce", dayfirst=True)
  return df
```

üîπ Aplicamos la normalizaci√≥n de texto en los distintos DataFrames ‚¨á:

```python
clientes_clean= normalizar_texto(clientes_clean)
ventas_clean= normalizar_texto(ventas_clean)
marketing_clean= normalizar_texto(marketing_clean)
```

# 5- Transformaci√≥n de datos: Aplicar filtros y transformaciones para crear una tabla de ventas que muestre solo los productos con alto rendimiento.

## 5.1 Calculamos el ingreso por registro = precio * cantidad üí∞:

```python
ventas_clean["ingreso"] = ventas_clean["precio"] * ventas_clean["cantidad"]
print(ventas_clean.head(5))
```

## 5.2 Agrupamos las m√©tricas por producto:

```python
resumen_prod = (
    ventas_clean.groupby(by="producto", dropna=False, as_index=False, observed=False)
    .agg(ingreso_total=('ingreso', 'sum'), unidades=('cantidad', 'sum'), precio_promedio=('precio', 'mean'), registros=('ingreso', 'size'))
)

print(resumen_prod.head(5))
```

## 5.3 Calculamos el percentil 80 de ingreso_total:

```python
p80_ingreso = resumen_prod["ingreso_total"].quantile(
    q=0.80,
    interpolation="linear"
)
```

## 5.4 Filtramos y ordenamos los productos que superen el p80_ingreso:

```python
ventas_top = (
    resumen_prod.query(
        "ingreso_total >= @p80_ingreso",
        engine="python")
    .sort_values(
        by=["ingreso_total", "unidades"],
        ascending=[False, False],
        na_position="last",
        ignore_index=False
    )
)
```

Mostramos los resultados finales ü•á:

```python
print(f"Umbral (percentil 80) de ingreso_total: {float(p80_ingreso):,.2f}")
print("\n‚úÖ Productos de ALTO RENDIMIENTO (top 20% por ingreso):\n")
display(ventas_top.head(20))
```

# 6- Agregaci√≥n: Resumir las ventas por categor√≠a de producto y analizar los ingresos generados.

Nos aseguramos que est√© la columna "ingreso" y si no existe, la creamos

```python
if "ingreso" not in ventas_clean.columns:
    ventas_cat = ventas_clean.assign(ingreso = ventas_clean["precio"] * ventas_clean["cantidad"])
else:
    ventas_cat = ventas_clean.copy()
```

Agrupamos por categor√≠a

```python
resumen_cat = (
    ventas_cat
    .groupby(
        by="categoria",
        dropna=False,
        as_index=False
    )
    .agg(
        ingreso_total=('ingreso', 'sum'),
        unidades=('cantidad', 'sum'),
        ventas=('ingreso', 'size'),
        precio_promedio=('precio', 'mean')
    )
    .sort_values(
        by='ingreso_total',
        ascending=False,
        na_position='last',
        ignore_index=True
    )
)
```

Calculamos el ticket promedio por venta

```python
resumen_cat['ticket_promedio_por_venta']= resumen_cat['ingreso_total'] / resumen_cat['ventas']
```

üîπ Mostramos los productos por categoria y sus ingresos:

```python
print("Resumen por categor√≠a (ordenado por ingreso_total):")
display(resumen_cat.head(20))
```

# 8- Estad√≠stica Descriptiva: Calcular medidas de tendencia central y dispersi√≥n sobre las ventas. Presentar los resultados en un informe.

Comenzaremos realizando un gr√°fico de barras:

üîπ Utilizamos una copia del DF resumen_prod que usamos anteriormente y ordenamos de manera ascendente.

```python
df = resumen_prod.copy()
df_ordenado = df.sort_values("ingreso_total", ascending=False)
plt.figure(figsize=(12, 6))
color_azul = "#1e81b0"
barras = plt.bar(df_ordenado["producto"], df_ordenado["ingreso_total"], color=color_azul)
plt.title("Ingresos Totales por Producto", fontsize=14, weight="bold")
plt.xlabel("Producto", fontsize=12)
plt.ylabel("Ingreso Total ($)", fontsize=12)
plt.xticks(rotation=90)

for barra in barras:
    y = barra.get_height()
    plt.text(
        barra.get_x() + barra.get_width() / 2,
        y / 2,
        f"${y:,.0f}",
        ha='center',
        va='center',
        fontsize=9,
        color="white",
        rotation=90
    )

plt.tight_layout()
plt.show()
```

üîπ Calculamos las medidas de tendencia central: media, mediana, moda.

```python
media = df['ingreso_total'].mean()
mediana = df['ingreso_total'].median()
moda = df['ingreso_total'].mode()

print("Media:", media)
print("Mediana:", mediana)
print("Moda:")
print(moda)
```

üîπ Calculamos las medidas de dispersi√≥n: rango, varianza, desviaci√≥n.

```python
rango = df['ingreso_total'].max() - df['ingreso_total'].min()
varianza = df['ingreso_total'].var(ddof=1)
desviacion = df['ingreso_total'].std(ddof=1)

print("Rango:", rango)
print("Varianza:", varianza)
print("Desviaci√≥n est√°ndar:", desviacion)
```

Resumen:

```python
print(f"Media de Ingreso Total: {media:,.2f}")
print(f"Mediana de Ingreso Total: {mediana:,.2f}")
print(f"Desviaci√≥n Est√°ndar de Ingreso Total: {desviacion:,.2f}")
print(f"Rango de Ingreso Total: {rango:,.2f}")
```

*   Media: 48,903.12
*   Mediana: 48,140.24
*   Desviaci√≥n Est√°ndar: 13,265.56
*   Rango de Ingreso Total: 71,147.58


**Media vs. Mediana:**

La media y la mediana son muy similares (diferencia menor al 2%), lo que indica que:
La distribuci√≥n de ingresos por producto es bastante equilibrada.
No hay productos extremadamente altos o bajos que est√©n sesgando la media,
no hay outliers fuertes que distorsionen la distribuci√≥n.

**Desviaci√≥n est√°ndar:**

Este valor representa la variabilidad respecto a la media.
Una desviaci√≥n est√°ndar de 13,265.56 representa aproximadamente 27% del promedio, lo cual indica una variabilidad significativa.

**Rango:**

Esto muestra la diferencia entre el producto que m√°s ingresa y el que menos.
Es un rango amplio, pero considerando la media (48,903.12), indica que hay productos que sobresalen y otros rezagados, aunque no al punto de generar sesgos muy fuertes.

**Conclusi√≥n:**

Aunque la media y la mediana son similares, lo que indica una distribuci√≥n sin sesgos extremos, la desviaci√≥n est√°ndar del 27% revela una variabilidad considerable entre los productos. Algunos productos generan ingresos significativamente superiores al promedio, mientras que otros se encuentran por debajo, lo que refleja un portafolio heterog√©neo. Este nivel de dispersi√≥n sugiere oportunidades de optimizaci√≥n en los productos de menor desempe√±o y una posible estrategia de refuerzo para los productos l√≠deres.









üîπ Calculamos el rango intercuart√≠lico (IQR) para la detecci√≥n de outliers

```python
q1 = df['ingreso_total'].quantile(0.25)
q2 = df['ingreso_total'].quantile(0.5)
q3 = df['ingreso_total'].quantile(0.75)
iqr = q3 - q1

print("Q1:", q1)
print("Mediana (Q2):", q2)
print("Q3:", q3)
print("IQR:", iqr)
```

üîπ Detecci√≥n de outliers

```python
lim_inf = q1 - 1.5 * iqr
lim_sup = q3 + 1.5 * iqr

outliers = df[(df['ingreso_total'] < lim_inf) | (df['ingreso_total'] > lim_sup)]

print("L√≠mite inferior:", lim_inf)
print("L√≠mite superior:", lim_sup)
print("\nOutliers detectados:")
print(outliers)
```
